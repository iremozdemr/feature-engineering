-min max 


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.width', 500)
```

```python
df = pd.read_csv("datasets/titanic.csv")
df.head()
```

# standard

```python
ss = StandardScaler()
df["Age_standard_scaler"] = ss.fit_transform(df[["Age"]])
print(df.head())
df.describe().T
```

# robust

```python
rs = RobustScaler()
df["Age_robust_scaler"] = rs.fit_transform(df[["Age"]])
print(df.head())
df.describe().T
```

# min max

```python
mms = MinMaxScaler()
df["Age_min_max_scaler"] = mms.fit_transform(df[["Age"]])
print(df.head())
df.describe().T
```

# grafikler

```python
# grafiklerle leklendirmeleri karlatrma
plt.figure(figsize=(15, 5))

# orijinal Age dalm
plt.subplot(1, 4, 1)
plt.hist(df["Age"].dropna(), bins=20, edgecolor="k", alpha=0.7)
plt.title("orijinal age dalm")
plt.xlabel("age")
plt.ylabel("Frequency")

# StandardScaler ile leklendirilmi Age dalm
plt.subplot(1, 4, 2)
plt.hist(df["Age_standard_scaler"].dropna(), bins=20, edgecolor="k", alpha=0.7)
plt.title("standard scaler")
plt.xlabel("standard scaled age")

# RobustScaler ile leklendirilmi Age dalm
plt.subplot(1, 4, 3)
plt.hist(df["Age_robust_scaler"].dropna(), bins=20, edgecolor="k", alpha=0.7)
plt.title("robust scaler")
plt.xlabel("robust scaled age")

# MinMaxScaler ile leklendirilmi Age dalm
plt.subplot(1, 4, 4)
plt.hist(df["Age_min_max_scaler"].dropna(), bins=20, edgecolor="k", alpha=0.7)
plt.title("min max scaler")
plt.xlabel("min max scaled age")

plt.tight_layout()
plt.show()
```
