```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.width', 500)
```

```python
df = pd.read_csv("datasets/application_train.csv")
df.head()
```

# label encoding

```python
le = LabelEncoder()
```

```python
le.fit_transform(df["FLAG_OWN_REALTY"])[0:5]
#alfabetik sıraya göre 0-1 verir
```

```python
le.inverse_transform([0,1])
#0-1 değerlerinin karşılığını verir
```

```python
def label_encoder(dataframe,columns):
    for column in columns:   
        labelencoder = LabelEncoder()
        dataframe[column] = labelencoder.fit_transform(dataframe[column])
        return dataframe
```

```python
binary_columns = [column for column in df.columns if df[column].dtype not in [int,float] and df[column].nunique() == 2]
#iki sınıflı kategorik değişkenleri bulma

print(binary_columns)
print(df[binary_columns].head())
```

```python
label_encoder(df,binary_columns)
print(df.head())
```

# one hot encoding

```python
df = pd.read_csv("datasets/titanic.csv")
df.head()
```

```python
value_count = df["Embarked"].value_counts()
print("value count: ",value_count)

na_count = df['Embarked'].isna().sum()
print("null count: ",na_count)
```

```python
df = pd.get_dummies(df,columns=["Embarked"],drop_first=True,dummy_na=True)
df.head()
#true-false yapıldı
#drop-first ile sütunlardan biri silinir
#dummy_na ile eksik değerler için de bir sütun oluşturulur
```

```python
dummy_columns = ["Embarked_Q","Embarked_S","Embarked_nan"]
df[dummy_columns] = df[dummy_columns].astype(int)
df.head()
#1-0 yapıldı
```

# rare encoding

```python
df = pd.read_csv("datasets/application_train.csv")
df.head()
```

```python
df["NAME_EDUCATION_TYPE"].value_counts()
```

```python
threshold = 0.2

rare_categories = df["NAME_EDUCATION_TYPE"].value_counts(normalize=True)
rare_categories = rare_categories[rare_categories < threshold].index

df["NAME_EDUCATION_TYPE"] = df["NAME_EDUCATION_TYPE"].replace(rare_categories,"Rare")

```
